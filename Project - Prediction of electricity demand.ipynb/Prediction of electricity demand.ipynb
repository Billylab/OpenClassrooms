```r
# multiplot : Plotting multiple graphs

multiplot <- function(..., plotlist=NULL, file, cols=1, layout=NULL) {
  library(grid)

  # Make a list from the ... arguments and plotlist
  plots <- c(list(...), plotlist)

  numPlots = length(plots)

  # If layout is NULL, then use 'cols' to determine layout
  if (is.null(layout)) {
    # Make the panel
    # ncol: Number of columns of plots
    # nrow: Number of rows needed, calculated from # of cols
    layout <- matrix(seq(1, cols * ceiling(numPlots/cols)),
                    ncol = cols, nrow = ceiling(numPlots/cols))
  }

 if (numPlots==1) {
    print(plots[[1]])

  } else {
    # Set up the page
    grid.newpage()
    pushViewport(viewport(layout = grid.layout(nrow(layout), ncol(layout))))

    # Make each plot, in the correct location
    for (i in 1:numPlots) {
      # Get the i,j matrix positions of the regions that contain this subplot
      matchidx <- as.data.frame(which(layout == i, arr.ind = TRUE))

      print(plots[[i]], vp = viewport(layout.pos.row = matchidx$row,
                                      layout.pos.col = matchidx$col))
    }
  }
}
```

## Verification of validity assumptions of linear regression

### Assessment of independence

```r
# Save figure
png("figure/Figure 6 - Lag plot : Residuals.png", width = 16, height = 9, units = 'in', res = 300)
lag.plot(regr$residuals, main="Figure 6 - Lag plot : Residuals")
dev.off()

# Plotting figure
lag.plot(regr$residuals, main="Figure 6 - Lag plot : Residuals")
```

```r
library(car)
durbinWatsonTest(regr)
```

<div class="alert alert-warning" style="text-align: justify">
The independence or not of the residues can be represented with the lag plot which represents the value of the residue n + 1 as a function of the value of the residue n. To have a dependence between the residues, a well-defined form must be observed, which is not the case.

Moreover, the Durbin Watson statistic is close to 2 p-value> 5%, which is characteristic of an independence of residues: the hypothesis of independence of the residues is thus verified.
</div>

### Hypothesis of normality of residues

<div class="alert alert-success" style="text-align: justify">
This hypothesis can be evaluated graphically using a QQplot. If the residues are well distributed along the line on the plot, then the assumption of normality is accepted. Conversely, if they deviate, then the hypothesis of normality is rejected
</div>

```r
# Save figure
png("figure/Figure 7 - QQ plot : Residuals.png", width = 16, height = 9, units = 'in', res = 300)
plot(regr,2)
dev.off()

# Plotting figure
plot(regr,2)
```

<div class="alert alert-warning" style="text-align: justify">
Here, the points are pretty well aligned on the line.

The Shapiro-Wilk test can also be used to assess the normality of residues. The normality assumption is rejected if the p-value is less than 0.05.
</div>

```r
shapiro.test(regr$residuals)
```

<div class="alert alert-warning" style="text-align: justify">
We have a p-value> 5% which leads us to accept the hypothesis H0: the residues are normally distributed.
</div>

### Residue homogeneity hypothesis

<div class="alert alert-success" style="text-align: justify">
This hypothesis can be verified in a visual way. It is necessary to realize a "residuals vs fitted plot". The "fitted" correspond to the responses predicted by the model, for the observed values of the predicitive variable.
More precisely, for this plot, the square root of the standardized residues is used.
</div>

```r
# Save figure
png("figure/Figure 8 - Scale-Location : Residuals.png", width = 16, height = 9, units = 'in', res = 300)
plot(regr, 3)
dev.off()

# Plotting figure
plot(regr, 3)
```

<div class="alert alert-warning" style="text-align: justify">
Here, the red curve is generally flat. This shows that the residuals tend to be homogeneously distributed throughout the gradient of predicted prestige values. And so the hypothesis of homogeneity of residues is accepted.

It is also possible to evaluate this hypothesis using the Breush-Pagan test. The homogeneity assumption is rejected if the p-value is less than 0.05.
</div>

```r
ncvTest(regr)
```

<div class="alert alert-warning" style="text-align: justify">
Here, the test does not reject the assumption of homogeneity either.
</div>

## <font color=#FEB229> 1. Holt Winters method (exponential smoothing)</font> <a name="31"></a>

### Modeling

<div class="alert alert-success" style="text-align: justify">
The Holt-Winters method, sometimes still used in practice, consists in supposing that Xt is approximate near T by aT + (t-T) bT + ST, where S is the period of the seasonal cycle of the time series.
</div>

```r
hw=ets(consumption_corr,model="AAA")
```

<div class="alert alert-warning" style="text-align: justify">
The model used 'AAA' allows to decompose the time series into an error (1st A) + a trend (2nd A) + a seasonality (3rd A).
A: for "Additive"
</div>

```r
summary(hw)
```

<div class="alert alert-warning" style="text-align: justify">
As alpha> gamma> beta, what has the most influence in the model of Holt Winters for realization at a time T is the previous moment, then the seasonality, then the trend.
</div>

### Prediction using the Holt Winters model over one year

```r
pred_model_hw=forecast(hw,h=12,level=95)
pred_hw=pred_model_hw$mean
pred_hw_l=ts(pred_model_hw$lower,start=c(2018,6),frequency=12)
pred_hw_u=ts(pred_model_hw$upper,start=c(2018,6),frequency=12)

# Save figure
png("figure/Figure 14 - Forecast : Holt Winters.png", width = 8, height = 9, units = 'in', res = 300)
ts.plot(consumption_corr,pred_hw,pred_hw_l,pred_hw_u,xlab="t",ylab="Corrected consumption",col=c(1,2,3,3),lty=c(1,1,2,2),lwd=c(1,3,2,2))
legend("topleft",legend=c("int95%_inf","int95%_sup"),col=c(3,3),lty=c(2,2),lwd=c(2,2))
title("Figure 14 - Forecast : Holt Winters")
dev.off()

# Plot figure
ts.plot(consumption_corr,pred_hw,pred_hw_l,pred_hw_u,xlab="t",ylab="Corrected consumption",col=c(1,2,3,3),lty=c(1,1,2,2),lwd=c(1,3,2,2))
legend("topleft",legend=c("int95%_inf","int95%_sup"),col=c(3,3),lty=c(2,2),lwd=c(2,2))
title("Figure 14 - Forecast : Holt Winters")
```

```r
# Sauvegarder la figure
png("figure/Figure 15 - Forecast : Holt Winters Zoom.png", width = 16, height = 9, units = 'in', res = 300)
ts.plot(window(consumption_corr,start=c(2017,5)),pred_hw,pred_hw_l,pred_hw_u,xlab="t",ylab="Corrected consumption",col=c(1,2,3,3),lty=c(1,1,2,2),lwd=c(1,3,2,2))
legend("topleft",legend=c("int95%_inf","int95%_sup"),col=c(3,3),lty=c(2,2),lwd=c(2,2))
title("Figure 15 - Forecast : Holt Winters Zoom")
dev.off()

# Afficher la figure
ts.plot(window(consumption_corr,start=c(2017,5)),pred_hw,pred_hw_l,pred_hw_u,xlab="t",ylab="Corrected consumption",col=c(1,2,3,3),lty=c(1,1,2,2),lwd=c(1,3,2,2))
legend("topleft",legend=c("int95%_inf","int95%_sup"),col=c(3,3),lty=c(2,2),lwd=c(2,2))
title("Figure 15 - Forecast : Holt Winters Zoom")
```

## <font color=#FEB229> 2. SARIMA method</font> <a name="32"></a>

<div class="alert alert-success" style="text-align: justify">
Our time series Consumption has a trend and seasonality so it depends on time. So it can not be modeled by an ARMA stationary process.

We will therefore stationarize the time series by differentiation.
</div>

### Stationnarisation

```r
# Save figure
png("figure/Figure 17 - ACF : Xt.png", width = 16, height = 9, units = 'in', res = 300)
Acf(consumption_corr,lag.max=36,ylim=c(-1,1),main=TeX("Figure 17 - ACF : X_t$"))
dev.off()

# Plot figure
Acf(consumption_corr,lag.max=36,ylim=c(-1,1),main=TeX("Figure 17 - ACF : X_t$"))
```

<div class="alert alert-warning" style="text-align: justify">
By observing the simple autocorrelogram of the time series consumption:
<li> Simple autocorrelations ρ (h) are neither close to 1 nor close to each other. Thus we do not differentiate in trend.</li>
<li> The ACF output of the series shows a slow decrease towards 0 for the multiples of 12. Similar behaviors of simple autorrelations ρ (h) are observed periodically. Indeed, ρ (12) and ρ (24) are quite close, there is a strong correlation / relationship between separate implementations of 12 months. We then perform a differentiation in "seasonality" with s = 12 : differentiation (I-B ^ 12).</li>
</div>

```r
consumption_corr_dif_12= diff(consumption_corr,lag=12,differences=1)

# Save figure
png("figure/Figure 18 - ACF : (I-B^12)Xt.png", width = 8, height = 4.5, units = 'in', res = 300)
Acf(consumption_corr_dif_12,lag.max=36,ylim=c(-1,1),main=TeX("Figure 18 - ACF : $(I-B^{12})X_t$"))
dev.off()

# Plot figure
Acf(consumption_corr_dif_12,lag.max=36,ylim=c(-1,1),main=TeX("Figure 18 - ACF : $(I-B^{12})X_t$"))
```

<div class="alert alert-warning" style="text-align: justify">
The ACF output of the seasonally differentiated series seems to be interpreted as a simple empirical autocorrelogram. We will therefore identify an ARMA model on the series (I-B ^ 12) Xt.
</div>

<div class="alert alert-success" style="text-align: justify">
On the other hand, the ndiffs () and nsdiffs () functions of R give the best number of differentiations to be made, respectively in trend and seasonality.
</div>

```r
d = ndiffs(consumption_corr)
D = nsdiffs(consumption_corr)

cat('d : Optimal number of differentiations in tendency to perform :', d,'\n')
cat('D : Optimal number of seasonal differentiations to be made :', D)
```

### Model identification, estimation and validation

<div class="alert alert-success" style="text-align: justify">
At the end of the stationarization, we obtain a stationary process. There is a theoretical result, called Wold decomposition, which tells us that any stationary process can be modeled by an ARMA process. We can devote ourselves to the potential choices of the polynomial coefficients (p, q, P and Q) by relying respectively on the partial autocorrelogram and the simple autocorrelogram estimated:
</div>

<div class="alert alert-warning" style="text-align: justify">
Determination of AR (p)

If our differentiated time series (in trend and / or seasonality) is a process AR (p), then a realization X_t only depends on X_ (t-1), X_ (t-2), ... X_ (tp) realizations preceding.
</div>

<img src="image/ARp.png">

<div class="alert alert-warning" style="text-align: justify">
So X_t has no relation with the realizations X_ (t-p-1) ... X_1. Thus, on the partial autocorrelogram, from the rank p + 1, the r (h) are zero.

The r (h) being the autocorrelations between time t and t-h conditionally at intermediate times.
</div>

<div class="alert alert-warning" style="text-align: justify">
** Determination of MA (q) **

If our differentiated time series (in trend and / or seasonality) is a MA (q) process, then an X_t realization can be written as a linear combination of decorrelated disturbances: a white noise and its past.
</div>

<img src="image/MAq.png">

<div class="alert alert-warning" style="text-align: justify">
So X_t has no connection with the variations of the realizations X_ (t-q-1) ... X_1. Thus, on the simple autocorrelogram, from the rank q + 1, the ρ (h) are zero.
</div>

### Determination of ARs (P) and MAs (Q)

<div class="alert alert-success" style="text-align: justify">
The seasonal part of the AR or MA models can be observed by looking at the seasonal lags (12,24 ...) of the ACP and the PACF.    
    <li> For example: an ARIMA model (0,0,0) (0,0,1) [12] will have: </li>
    <ul>
    <li> a peak at lag 12 in the ACF but no other significant peaks</li>
    <li> a decrease in seasonal lags in the PACF (ie lags 12, 24, 36 ...)</li> 
    </ul>
    <li> For example: an ARIMA model (0,0,0) (1,0,0) [12] will have: </li>
    <ul>
    <li> a decrease in seasonal lags in the ACF (ie lags 12, 24, 36 ...)</li>
    <li> a peak at lag 12 in the PACF but no other significant peaks</li> 
    </ul>
</div>

```r
# Plot figure
Acf(consumption_corr_dif_12,lag.max=36,ylim=c(-1,1),main=TeX("Figure 18 - ACF : $(I-B^{12})X_t$"))
```

```r
# Save figure
png("figure/Figure 19 - PACF : (I-B^12)Xt.png", width = 8, height = 4.5, units = 'in', res = 300)
Pacf(consumption_corr_dif_12,lag.max=36,ylim=c(-1,1),main=TeX("Figure 19 - PACF : $(I-B^{12})X_t$"))
dev.off()

# Plot figure
Pacf(consumption_corr_dif_12,lag.max=36,ylim=c(-1,1),main=TeX("Figure 19 - PACF : $(I-B^{12})X_t$"))
```

<div class="alert alert-warning" style="text-align: justify">
The blue dashed lines are the boundaries of significance (95%).

We have :   
    <li> p = 0 because: from lag 1, the partial autocorrelations are not significant </li>
    <li> P = 1 because: </li>
        <ul>
        <li> a decrease in seasonal lags in the ACF (ie lags 12, 24, 36 ...)</li>
        <li> a peak at lag 12 in the PACF but no other significant peaks</li> 
        </ul>
    <li> q = 0 because: from lag 1, simple autocorrelations are not significant </li>
    <li> Q = 1 because: </li>
        <ul>
        <li> a peak at lag 12 in the ACF but no other significant peaks</li>
        <li> a decrease in seasonal lags in the PACF (ie lags 12, 24, 36 ...)</li> 
        </ul>
</div>
