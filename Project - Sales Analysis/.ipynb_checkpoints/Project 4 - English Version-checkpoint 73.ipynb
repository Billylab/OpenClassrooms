### Cleaning

```python
# Test values are removed
customers = customers[customers['client_id']!='ct_0']
customers = customers[customers['client_id']!='ct_1']

products = products[products['id_prod']!='T_0']

transactions = transactions[transactions['id_prod']!='T_0']
transactions = transactions[transactions['client_id']!='ct_0']
transactions = transactions[transactions['client_id']!='ct_1']
transactions = transactions[transactions['session_id']!='s_0']
```

### Observations

```python
nb_customers = customers.shape[0]
nb_products = products.shape[0]
nb_transactions = transactions.shape[0]
print(nb_customers," customers registered on the website.")
print(nb_products," produits registered on the website.")
print(nb_transactions," sales done on the website.")
```

## <font color=#FEB229> b. Data : Products, transactions and customers</font> <a name="2b"></a>

```python
# Customer ages are added in the customers table
customers['age'] = 2022 - customers['birth']
customers.head()
```

```python
# data : outer join of products & transactions on id_prod : keep all products & transactions
data = pd.merge(products, transactions, how='outer', left_on = 'id_prod', right_on = 'id_prod')

# data : outer join of customers & data on client_id : keep all produits, transactions & customers
data = pd.merge(customers, data, how='outer', left_on = 'client_id', right_on = 'client_id')

data = data.rename(columns={"date":"datetime"})

# date : Date Purchased
data['date'] = pd.DatetimeIndex(data['datetime']).date

# year : Year
data['year'] = pd.DatetimeIndex(data['datetime']).year

# month : Month
data['month'] = pd.DatetimeIndex(data['datetime']).month

# yearmonth : Year & Month
data['yearmonth'] = pd.DatetimeIndex(data['datetime']).map(lambda x: x.strftime('%Y-%m'))

# day : Day
data['day'] = pd.DatetimeIndex(data['datetime']).day

# Sorted Data by datetime
data = data.sort_values(by=['datetime'])

data.head()
```

## <font color=#FEB229> d. Analysis of the loss of sales revenue in October</font> <a name="2d"></a>

```python
# x_start : Past days since the beginning
x_start = data.groupby(["date"])["date"].first()

# y_start : Sales revenue per day since the beginning, for products from category 0
y_start = data.groupby(["date"])["price"].sum()

# curve_start : Dataframe 
curve_start = pd.DataFrame([a for a in zip(y_start,x_start)])
curve_start.columns = ["price","date"]

# Tracé de la figure
plt.figure(figsize = (31, 17))
plt.plot(curve_start.date, curve_start.price, 'b',linewidth=6)
plt.xlabel('Date',fontsize=30,fontweight = 'bold')
plt.ylabel('Sales Revenue',fontsize=30,fontweight = 'bold')
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.xticks(rotation = 'vertical')
plt.grid(True,linestyle='--')
plt.title('Figure 1 - Sales revenue evolution',fontsize=40,fontweight = 'bold')
plt.savefig('figures/Figure 1 - Sales revenue evolution.png', dpi=300)
plt.show()
plt.close()
```

<div class="alert alert-warning">
A significant decrease occured during october 2021 !

```python
# october : Data in October
october = data[data['month']==10]

# day_october : Days of october ranked in ascending order
day_october = october.groupby(["day"])["day"].first()

# day_october1 : Days of october ranked in ascending order, for products from category 1
day_october1 = october[october['categ']==1].groupby(["day"])["day"].first()

# y0 : Sales revenue per day in October for products from category 0
y0 = october[october['categ']==0].groupby(["day"])["price"].sum()
y1 = october[october['categ']==1].groupby(["day"])["price"].sum()
y2 = october[october['categ']==2].groupby(["day"])["price"].sum()

# courbe0 : Dataframe : Sales revenue per day in October for products from category 0
df0 = pd.DataFrame([a for a in zip(day_october,y0)])
df0.columns = ["date","price0"]

df1 = pd.DataFrame([a for a in zip(day_october1,y1)])
df1.columns = ["date","price1"]

df2 = pd.DataFrame([a for a in zip(day_october,y2)])
df2.columns = ["date","price2"]

# Plotting figure
plt.figure(figsize = (31, 17))
plt.plot(df0.date, df0.price0, 'b', label='categ 0',linewidth=6)
plt.scatter(df1.date, df1.price1.values, c='g', label='categ 1',linewidth=6)
plt.plot(df2.date, df2.price2, 'r', label='categ 2',linewidth=6)
plt.xlabel('Days in October',fontsize=30,fontweight = 'bold')
plt.ylabel('Sales revenue evolution in October',fontsize=30,fontweight = 'bold')
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.legend(loc='lower right',fontsize=30)
plt.xticks(rotation = 'vertical')
plt.grid(True,linestyle='--')
plt.title('Figure 2 - Sales revenue evolution in October',fontsize=30,fontweight = 'bold')
plt.savefig('figures/Figure 2 - Sales revenue evolution in October.png', dpi=300)
plt.show()
plt.close()
```

<div class="alert alert-danger" role="alert">
  Between the 2nd and the 27th of October, <strong> there was no information feedback on products from category 1 </strong> : their prices are missing ! 
</div>

<div class="alert alert-success" role="alert" style="text-align: justify">
  I will make data imputation for missing data using multiple linear regression which computes a model for <strong> Sales revenue for category 1 </strong > expressed with <strong> Sales revenue for category 0 </strong > and <strong> Sales revenue for category 2 </strong >.
</div>

```python
# data_imputation : Data before imputation of prices for products from category 1 : before 10/02/2021
data_imputation = data[data['datetime']<='2021-10-02']

# x_imputation : Past days before 10/02/2021
x_imputation = data_imputation.groupby(["date"])["date"].first()

# y0_imputation : Sales revenue per day before 10/02/2021 for products from category 0
y0_imputation = data_imputation[data_imputation['categ']==0].groupby(["date"])["price"].sum()
y1_imputation = data_imputation[data_imputation['categ']==1].groupby(["date"])["price"].sum()
y2_imputation = data_imputation[data_imputation['categ']==2].groupby(["date"])["price"].sum()

# categorie0 : Dataframe : Sales revenue per day before 10/02/2021 for products from category 0
categorie0 = pd.DataFrame([a for a in zip(x_imputation,y0_imputation)])
categorie0.columns = ["date","price0"]

categorie1 = pd.DataFrame([a for a in zip(x_imputation,y1_imputation)])
categorie1.columns = ["date","price1"]

categorie2 = pd.DataFrame([a for a in zip(x_imputation,y2_imputation)])
categorie2.columns = ["date","price2"]

# data_imputation : left join of categorie0 & categorie1 on date 
data_imputation = pd.merge(categorie0, categorie1, on = ['date'],how='left')

# data_imputation : left join of data_imputation & categorie2 on date
data_imputation = pd.merge(data_imputation, categorie2, on = ['date'],how='left')

data_imputation.head()
```

```python
# Plotting figure
plt.figure(figsize = (31, 17))
plt.plot(data_imputation.date, data_imputation.price0, 'b', label='categ 0',linewidth=6)
plt.plot(data_imputation.date, data_imputation.price1, c='g', label='categ 1',linewidth=6)
plt.plot(data_imputation.date, data_imputation.price2, 'r', label='categ 2',linewidth=6)
plt.xlabel('Days up to October 2021',fontsize=30,fontweight = 'bold')
plt.ylabel('Sales revenue up to October 2021',fontsize=30,fontweight = 'bold')
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.legend(loc='lower right',fontsize=30)
plt.xticks(rotation = 'vertical')
plt.grid(True,linestyle='--')
plt.title('Figure 3 - Sales revenue evolution up to October 2021',fontsize=30,fontweight = 'bold')
plt.savefig('figures/Figure 3 - Sales revenue evolution up to October 2021.png', dpi=300)
plt.show()
plt.close()
```

```python
# reg_multi : Multiple linear regression
reg_multi = smf.ols('price1~price0+price2', data=data_imputation).fit()
print(reg_multi.summary())
```

<div class="alert alert-warning" style="text-align: justify">
$R^{2}$ =  0.145 : Our model weakly explains the variation. To explain the sales revenue for products from category 1, our variables are not enough.
The coefficients of price0 and price2 are significant with a 95% confidence interval because their P-values are less than 0.05.

Despite a low correlation coefficient, it is a fairly clean imputation of the missing data.
</div>

```python
# imputation : Dataframe : Sales revenue per day in October per category
imputation = pd.merge(df0, df1, on = ['date'],how='left')
imputation = pd.merge(imputation, df2, on = ['date'],how='left')
imputation.head()
```

```python
# Multiple Linear Regression : Results
b = reg_multi.params.Intercept
a0 = reg_multi.params.price0
a2 = reg_multi.params.price2

# Missing-data imputation
for i in range(1,27):
    imputation.price1[i] = a0*imputation.price0[i] + a2*imputation.price2[i] + b
```

```python
# date_imputation : Days of data imputation 
date_imputation = [i for i in range (1,29)]

# price1_imputation : Imputed values of sales revenue per day for category 1
price1_imputation = [imputation.price1[i] for i in range(0,29)]

# df1_imput : Dataframe : Imputed values of sales revenue per day for category 1
df1_imput = pd.DataFrame([a for a in zip(date_imputation,price1_imputation)])
df1_imput.columns = ["date","price1"]

# Plotting figure
plt.figure(figsize = (31, 17))
plt.plot(df0.date, df0.price0, 'b', label='categ 0',linewidth=6)
plt.scatter(df1.date, df1.price1.values, c='g', label='categ 1',linewidth=6)
plt.plot(df1_imput.date, df1_imput.price1, 'black', label='categ 1 imputée',linewidth=6)
plt.plot(df2.date, df2.price2, 'r', label='categ 2',linewidth=6)
plt.xlabel('Days in October',fontsize=30,fontweight = 'bold')
plt.ylabel('Sales revenue',fontsize=30,fontweight = 'bold')
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.legend(loc='lower right',fontsize=30)
plt.xticks(rotation = 'vertical')
plt.grid(True,linestyle='--')
plt.title('Figure 4 - Sales revenue evolution in October, after data imputation',fontsize=30,fontweight = 'bold')
plt.savefig('figures/Figure 4 - Sales revenue evolution in October, after data imputation.png', dpi=300)
plt.show()
plt.close()
```

### Analysis of the number of sales per month

```python
# nb_sales_month_012 : Number of sales per month
nb_sales_month_012 = data.yearmonth.value_counts().sort_index()
nb_sales_month_0 = data[data['categ']==0].yearmonth.value_counts().sort_index()
nb_sales_month_1 = data[data['categ']==1].yearmonth.value_counts().sort_index()
nb_sales_month_2 = data[data['categ']==2].yearmonth.value_counts().sort_index()

# nb_ventes_mois_012 : Dataframe : Number of sales per month
nb_sales_month_012 = pd.DataFrame([a for a in zip(x_month,nb_sales_month_012)])
nb_sales_month_0 = pd.DataFrame([a for a in zip(x_month,nb_sales_month_0)])
nb_sales_month_1 = pd.DataFrame([a for a in zip(x_month,nb_sales_month_1)])
nb_sales_month_2 = pd.DataFrame([a for a in zip(x_month,nb_sales_month_2)])

nb_sales_month_012.columns = ["month","nb_sales012"]
nb_sales_month_0.columns = ["month","nb_sales0"]
nb_sales_month_1.columns = ["month","nb_sales1"]
nb_sales_month_2.columns = ["month","nb_sales2"]
```

<div class="alert alert-danger" role="alert">
  Between the 2nd and the 27th of October, <strong> there was no information feedback on products from category 1 </strong> : their prices are missing ! 
</div>

<div class="alert alert-success" role="alert" style="text-align: justify">
  I will make data imputation for missing data using multiple linear regression which computes a model for <strong> Number of sales of category 1 </strong > expressed with <strong> Number of sales of category 0 </strong > and <strong> Number of sales of category 2 </strong >.
</div>

```python
# data_imputation_nb_sales : Number of sales per category & per month, before October 2021
data_imputation_nb_sales = pd.merge(nb_sales_month_0, nb_sales_month_1, on = ['month'],how='left')
data_imputation_nb_sales = pd.merge(data_imputation_nb_sales, nb_sales_month_2, on = ['month'],how='left')
data_imputation_nb_sales = data_imputation_nb_sales[data_imputation_nb_sales['month']<='2021-09']
data_imputation_nb_sales
```

```python
# reg_multi_sales : Multiple linear regression
reg_multi_sales = smf.ols('nb_sales1~nb_sales0+nb_sales2', data=data_imputation_nb_sales).fit()
print(reg_multi_sales.summary())
```

<div class="alert alert-warning" style="text-align: justify">
$R^{2}$ =  0.786 : Our model explains the variation.
    
The coefficients of nb_sales0 and nb_sales2 are significant with a 95% confidence interval because their P-values are less than 0.05.

```python
# SR_client_list : Sales revenue per customer
SR_client_list = np.asarray(data.groupby(["client_id"])["price"].sum().tolist())
lorenz = np.cumsum(np.sort(SR_client_list)) / SR_client_list.sum()
lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0
x = np.linspace(0,1,len(lorenz))

# Plotting figure
plt.figure(figsize = (31, 17))
plt.plot(x,lorenz,'b',drawstyle='steps-post',linewidth=6, label='Lorenz curve')
plt.plot([0,1],[0,1],'red',linestyle='--',linewidth=6, label='Case Full Equity')
plt.fill_between(x, x, lorenz,facecolor='grey', label = 'Surface')
plt.xlabel('Proportion of buyers',fontsize=30,fontweight = 'bold')
plt.ylabel('Proportion of the sales revenue ',fontsize=30,fontweight = 'bold')
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.legend(loc='upper left',fontsize=30)
plt.grid(True,linestyle='--')
plt.title('Figure 10 - Lorenz curve : Customers and Sales revenue',fontsize=30,fontweight = 'bold')
plt.savefig('figures/Figure 10 - Lorenz curve : Customers and Sales revenue.png', dpi=300)
plt.show()
```

        <li> Statistic F: F = f (Variance Inter / Variance Intra): Follows a law of Fischer </li>
    </ul>
</div>

        <li> F: 687: Very high: The interclass variance is much larger than the intraclass variance. </li>
