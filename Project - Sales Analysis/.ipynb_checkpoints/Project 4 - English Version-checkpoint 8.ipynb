### Cleaning

```python
# Test values are removed
customers = customers[customers['client_id']!='ct_0']
customers = customers[customers['client_id']!='ct_1']

products = products[products['id_prod']!='T_0']

transactions = transactions[transactions['id_prod']!='T_0']
transactions = transactions[transactions['client_id']!='ct_0']
transactions = transactions[transactions['client_id']!='ct_1']
transactions = transactions[transactions['session_id']!='s_0']
```

### Observations

## <font color=#FEB229> b. Data : Products, transactions and customers</font> <a name="2b"></a>

```python
# Customer ages are added in the customers table
customers['age'] = 2022 - customers['birth']
customers.head()
```

```python
# data : outer join of products & transactions on id_prod : keep all products & transactions
data = pd.merge(products, transactions, how='outer', left_on = 'id_prod', right_on = 'id_prod')

# data : outer join of customers & data on client_id : keep all produits, transactions & customers
data = pd.merge(customers, data, how='outer', left_on = 'client_id', right_on = 'client_id')

data = data.rename(columns={"date":"datetime"})

# date : Date Purchased
data['date'] = pd.DatetimeIndex(data['datetime']).date

# year : Year
data['year'] = pd.DatetimeIndex(data['datetime']).year

# month : Month
data['month'] = pd.DatetimeIndex(data['datetime']).month

# yearmonth : Year & Month
data['yearmonth'] = pd.DatetimeIndex(data['datetime']).map(lambda x: x.strftime('%Y-%m'))

# day : Day
data['day'] = pd.DatetimeIndex(data['datetime']).day

# Sorted Data by datetime
data = data.sort_values(by=['datetime'])

data.head()
```

```python
# data_imputation : Data before imputation of prices for products from category 1 : before 10/02/2021
data_imputation = data[data['datetime']<='2021-10-02']

# x_imputation : Past days before 10/02/2021
x_imputation = data_imputation.groupby(["date"])["date"].first()

# y0_imputation : Sales revenue per day before 10/02/2021 for products from category 0
y0_imputation = data_imputation[data_imputation['categ']==0].groupby(["date"])["price"].sum()
y1_imputation = data_imputation[data_imputation['categ']==1].groupby(["date"])["price"].sum()
y2_imputation = data_imputation[data_imputation['categ']==2].groupby(["date"])["price"].sum()

# categorie0 : Dataframe : Sales revenue per day before 10/02/2021 for products from category 0
categorie0 = pd.DataFrame([a for a in zip(x_imputation,y0_imputation)])
categorie0.columns = ["date","price0"]

categorie1 = pd.DataFrame([a for a in zip(x_imputation,y1_imputation)])
categorie1.columns = ["date","price1"]

categorie2 = pd.DataFrame([a for a in zip(x_imputation,y2_imputation)])
categorie2.columns = ["date","price2"]

# data_imputation : left join of categorie0 & categorie1 on date 
data_imputation = pd.merge(categorie0, categorie1, on = ['date'],how='left')

# data_imputation : left join of data_imputation & categorie2 on date
data_imputation = pd.merge(data_imputation, categorie2, on = ['date'],how='left')

data_imputation.head()
```

```python
# Plotting figure
plt.figure(figsize = (31, 17))
plt.plot(data_imputation.date, data_imputation.price0, 'b', label='categ 0',linewidth=6)
plt.plot(data_imputation.date, data_imputation.price1, c='g', label='categ 1',linewidth=6)
plt.plot(data_imputation.date, data_imputation.price2, 'r', label='categ 2',linewidth=6)
plt.xlabel('Days up to October 2021',fontsize=30,fontweight = 'bold')
plt.ylabel('Sales revenue up to October 2021',fontsize=30,fontweight = 'bold')
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.legend(loc='lower right',fontsize=30)
plt.xticks(rotation = 'vertical')
plt.grid(True,linestyle='--')
plt.title('Figure 3 - Sales revenue evolution up to October 2021',fontsize=30,fontweight = 'bold')
plt.savefig('figures/Figure 3 - Sales revenue evolution up to October 2021.png', dpi=300)
plt.show()
plt.close()
```

```python
# reg_multi : Multiple linear regression
reg_multi = smf.ols('price1~price0+price2', data=data_imputation).fit()
print(reg_multi.summary())
```

<div class="alert alert-warning" style="text-align: justify">

Ici, les bo√Ætes √† moustaches semblent distinctes. La cat√©gorie d'√¢ges semble avoir un impact sur le montant total des achats. L'**ANOVA** nous permet de confirmer ou d'infirmer cette intuition.  

**R√©alisation d'un test de Fisher :**  
- Hypoth√®se H0 : La cat√©gorie d'√¢ge n'a pas d'impact sur le montant total des achats :   $Œ±_{1}=Œ±_{2}=Œ±_{3}=0$.
- Hypoth√®se H1 : La cat√©gorie d'√¢ge a un impact sur le montant total des achats.
- Seuil de significativit√© : Œ± = 5%  
- Statistique F : F = f(Variance Inter / Variance Intra) : Suit une loi de Fischer

```python
anova_categ_age_montant_total_achat = smf.ols('montant_total_achat~categ_age', data=correlation_montant_total_achat_age).fit()
print(anova_categ_age_montant_total_achat.summary())
```

Pour obtenir le tableau de l'analyse de la variance, on utilise la commande anova_lm :

```python
sm.stats.anova_lm(anova_categ_age_montant_total_achat, typ=2)
```

#### Analyse des r√©sultats de l'ANOVA

- $eta^{2}$ : 0.078 : Les cat√©gories d'√¢ges expliquent peu (√† 7.8%) la variabilit√© du montant total des achats.
- F : 687 : Tr√®s √©lev√© : La variance interclasse est beaucoup plus importante que la variance intraclasse.
- P-valeur : < 5% : L'hypoth√®se H0 selon laquelle $Œ±_{1}=Œ±_{2}=Œ±_{3}=0$ est rejett√©e.

Pour obtenir le tableau de l'analyse de la variance, on utilise la commande anova_lm :

```python
sm.stats.anova_lm(anova_age_categ, typ=2)
```

#### Analyse des r√©sultats de l'ANOVA

- $eta^{2}$ : 0.001 : La variabilit√© des donn√©es n'est pas expliqu√©e par l'√¢ge
- F : 237 : Elev√© : La variance interclasse est plus importante que la variance intraclasse.
- P-valeur : < 5% : L'hypoth√®se H0 selon laquelle $Œ±_{1}=Œ±_{2}=Œ±_{3}=0$ est rejett√©e.

<font color='red'>L'√¢ge a donc bien un effet sur la cat√©gorie du produit achet√©, comme nous en avions l'intuition en regardant les bo√Ætes √† moustaches.</font>

#### Qu'en est-il √† l'ext√©rieur des cat√©gories ? Y a-t-il une corr√©lation entre les cat√©gories d'√¢ges des clients et la cat√©gorie du produit achet√© ? 

```python
# [19-29 ans] [30-49 ans] [50-93 ans]
data_correlations = data_correlations.sort_values(by=['age'])
conditions = [
    (data_correlations['age'] <= 29),
    (data_correlations['age'] >= 30) & (data_correlations['age'] <= 49),
    (data_correlations['age'] >= 50)]
choices = ['- de 30 ans', '30-49 ans', '+ de 49 ans']
data_correlations['categ_age'] = np.select(conditions, choices)
data_correlations.head()
```

```python
X = "categ"
Y = "categ_age"

# categ_categ_age : Nombre de ventes par cat√©gorie de produits et par cat√©gorie d'√¢ges
categ_categ_age = data_correlations[[X,Y]].pivot_table(index=X,columns=Y,aggfunc=len)
categ_categ_age = categ_categ_age.fillna(0)
cont = categ_categ_age.copy()

# tx : Total des produits achet√© par cat√©gorie de produits
tx = data_correlations[X].value_counts()

# ty : Total des produits achet√© par cat√©gorie d'√¢ges
ty = data_correlations[Y].value_counts()

cont.loc[:,"Total"] = tx
cont.loc["total",:] = ty
cont.loc["total","Total"] = len(data_correlations)
cont.head()
```

```python
tx = pd.DataFrame(tx)
ty = pd.DataFrame(ty)
tx.columns = ["foo"]
ty.columns = ["foo"]
n = len(data_correlations)
indep = tx.dot(ty.T) / n

mesure = (categ_categ_age-indep)**2/indep
xi_n = mesure.sum().sum()
```

```python
# Trac√© de la figure
plt.subplots(figsize=(20,17))
ax = sns.heatmap(mesure/xi_n,annot=categ_categ_age,linewidths=0.5,annot_kws={"size": 20})
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("Cat√©gorie d √¢ges",fontsize=30,fontweight = 'bold')
plt.ylabel("Cat√©gorie du produit",fontsize=30,fontweight = 'bold')
plt.title('Figure 22 - Heatmap - Cat√©gorie d √¢ges et cat√©gorie du produit',fontsize=30,fontweight = 'bold')
plt.savefig('figure/Figure 22 - Heatmap - Cat√©gorie d √¢ges et cat√©gorie du produit.png',dpi=300)
plt.show()
```

```python
# Test d'ind√©pendance du Chi2
chi2, pvalue, degrees, expected = chi2_contingency(np.array(categ_categ_age))
print('chi2 :',chi2)
print('degrees :',degrees)
print('pvalue :',pvalue)
```
